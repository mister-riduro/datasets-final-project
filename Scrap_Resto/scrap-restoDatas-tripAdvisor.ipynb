```python
from decimal import Decimal
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
from ast import IsNot


restoName = []
restoAddress = []
restoPhoneNumber = []
restoCuisineType = []
restoRating = []
restoOpenHour = []
restoCloseHour = []
restoWebLink = []
lattitude = []
longitude = []


driver = webdriver.Chrome('D:/College/final-project/Datasets/Scrap_Resto/chromedriver.exe')

for i in range(0, len(link_resto)):
  url=link_resto[i]

  driver.get(url)
  req=requests.get(url, headers={'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"})

  soup=BeautifulSoup(req.text, 'html.parser')

  for a in soup.findAll('div', attrs={'class':'page'}):

    resto_name = a.find('h1', attrs={'class' : 'HjBfq'})


    resto_address = a.findAll('a', attrs = {'class' : 'AYHFM'})
    if len(resto_address) > 0:
      resto_address = resto_address[1].text
    else:
      resto_address = "-"
    
    
    resto_phone_number = a.findAll('a', attrs = {'class' : 'BMQDV _F G- wSSLS SwZTJ'})
    if len(resto_phone_number) > 1:
      resto_phone_number = resto_phone_number[1].text
    else:
      resto_phone_number = "-"
    
    resto_rating = a.find('span', attrs = {'class' : 'ZDEqb'})


    #GET a ATTRIBUTE
    a_attribute = driver.find_element('xpath','//a[@class="YnKZo Ci Wc _S C FPPgD"]')
    #GET RESTO LINK
    resto_link = a_attribute.get_attribute("href")


    currency = "Rp"
    resto_cuisine = a.findAll('div', attrs = {'class' : 'SrqKb'})
    if len(resto_cuisine) > 1:
      if currency in resto_cuisine[0].text:
        resto_cuisine = resto_cuisine[1].text
      else:
        resto_cuisine = resto_cuisine[0].text
    else:
      resto_cuisine = resto_cuisine[0].text

    # GET IMAGE LINK
    try:
      image_map = driver.find_element('xpath','//img[@class="w MD _S"]')

      map_link = image_map.get_attribute('src')

      #ADJUST IMAGE LINK
      getCor = map_link.split("|")[1].split("&")
      lat, long = getCor[0].split(",")
      lat = Decimal(lat)
      long = Decimal(long)
      lattitude.append(lat)
      longitude.append(long)
    except NoSuchElementException:
      lattitude.append(0.0)
      longitude.append(0.0)
      pass

    #===============================================#
    #INSERT TO ARRAY
    restoName.append(resto_name.text)
    restoAddress.append(resto_address)
    restoPhoneNumber.append(resto_phone_number)
    restoCuisineType.append(resto_cuisine)
    restoRating.append(resto_rating.text)
    restoWebLink.append(resto_link)
```

```python
df = pd.DataFrame(list(zip(restoName, restoAddress, restoPhoneNumber, restoCuisineType, restoRating, restoWebLink, lattitude, longitude)), columns =['Resto Name', 'Address', 'Resto Phone Number', 'Tipe Masakan','Resto Rating', 'Link Website', 'Lattitute', 'Longitude']) 
df
```
