```python
from decimal import Decimal
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
room_price = ''
tempStr = ''
maxPrice = []
minPrice = []
cityHotel = []
addressHotel = []
hotelName = []
facilities = []
hotelRating = []
lattitude = []
longitude = []
hotelImage = []
propertyType = []
wifi = 'wifi'


for i in range(0, len(link_hotel)):
  priceHotel = []
  url=link_hotel[i]

  # print(url)
  req=requests.get(url)
  content=req.text

  soup=BeautifulSoup(content)

  for a in soup.findAll('div', attrs={'class':'hotel-detail__container'}):

    room_price = a.findAll('div', attrs={'class':'room-price'})
    hotel_name = a.find('h1', attrs={'class':'property-name tix-font h4 line-clamp-2'})
    address_hotel = a.find('div', attrs={'class' : 'location-address'})
    city_hotel = a.find('div', attrs={'class' : 'location ellipsis'})
    hotel_rating = a.find('div', attrs={'class' : 'score tiket'})
    image_map = a.find('div', attrs={'class' : 'location-map'})
    image_hotel_div = a.find('div', attrs={'class' : 'image'})
    property_type = a.find('div', attrs={'class' : 'property-type'})

    # i = 0
    if len(room_price):
      hotelName.append(hotel_name.text)

      city_hotel = city_hotel.text.replace(',', '')
      cityHotel.append(city_hotel)

      # print(image_hotel_div)

      if image_hotel_div is None:
        image_hotel_default = a.find('div', attrs={'class' : 'photo-default'})
        image_hotel = image_hotel_default.find('img')
        image_hotel = image_hotel.get('src')
        image_hotel = "https://www.tiket.com/"+image_hotel
        hotelImage.append(image_hotel)
      else:
        image_hotel = image_hotel_div.find('img')
        image_hotel = image_hotel.get('src')
        hotelImage.append(image_hotel)

      address_hotel = address_hotel.find('div', attrs={'class' : 'line-clamp-2'})
      address_hotel = address_hotel.text.replace(',','.')
      addressHotel.append(address_hotel)

      propertyType.append(property_type.text)

      if hotel_rating is None:
        hotelRating.append("Belum ada rating")
      else:  
        hotelRating.append(hotel_rating.text)

      #GET IMAGE LINK
      image_map = image_map.find('img')
      map_link = image_map.get('src')

      #ADJUST IMAGE LINK
      lat, long = map_link.split("|")[2].split(",")

      lat = Decimal(lat)
      long = Decimal(long)

      lattitude.append(lat)
      longitude.append(long)

      # Facilities
      # facil = a.findAll('div', attrs={'class':'facility-label'})
      # for b, data in enumerate(facil):
      #   facil[b] = data.text
      # facilities.append(facil)

      for c in range(len(room_price)):
        tempStr = room_price[c].find('span').text
        tempStr = tempStr.strip()
        tempStr = tempStr.lstrip('IDR ')
        tempStr = tempStr.replace('.', '')
        price = int(tempStr)
        priceHotel.append(price)
      maxPrice.append(max(priceHotel))
      minPrice.append(min(priceHotel))

    else:
      continue
```

```python
# print(type(minPrice))

# df = pd.DataFrame(list(zip(hotelImage, hotelName, propertyType, cityHotel, addressHotel, hotelRating, minPrice, maxPrice,facilities, lattitude, longitude)), columns =['Hotel Image', 'Hotel Name', 'Tipe Properti','City', 'Address', 'Hotel Rating', 'Min', 'Max', 'Facilities', 'Lattitute', 'Longitude']) 
# df

df = pd.DataFrame(list(zip(hotelImage, hotelName, propertyType, cityHotel, addressHotel, hotelRating, minPrice, maxPrice, lattitude, longitude)), columns =['Hotel Image', 'Hotel Name', 'Tipe Properti','City', 'Address', 'Hotel Rating', 'Min', 'Max', 'Lattitute', 'Longitude']) 
df.index += 349

df

# len(minPrice)
```
