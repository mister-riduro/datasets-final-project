```python
facilities_id = []
hotel_id = []
id_hotel = 403

for i in range(0, len(link_hotel)):
  url=link_hotel[i]
  # print(url)
  req=requests.get(url)
  content=req.text

  soup=BeautifulSoup(content)

  for a in soup.findAll('div', attrs={'class':'hotel-detail__container'}):
    room_price = a.findAll('div', attrs={'class':'room-price'})

    if len(room_price):
      for b in soup.findAll('div', attrs={'class':'card pdp-card facility-list'}):

        dunno = b.find('div', attrs={'class' : 'group-content'})

        label_div = dunno.findAll('div', attrs={'class' : 'facility-label'})
        # print(label_div)
        for c in range(len(label_div)):
          for d in range (len(fasilitas)):
            if label_div[c].text == fasilitas [d]:
              hotel_id.append(id_hotel)

              facilities_id.append(d + 1)
      
      id_hotel += 1
```

```python
df = pd.DataFrame(list(zip(facilities_id, hotel_id)), columns =['list_fasil_id', 'hotel_id']) 
df
```
